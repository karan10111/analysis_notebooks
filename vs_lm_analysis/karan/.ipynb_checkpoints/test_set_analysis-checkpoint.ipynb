```python
import pandas as pd
import seaborn as sns
```

```python
na_ol_night = pd.read_csv(filepath_or_buffer="/shared/suryansh.tiwari/output_csv/HI_FA_EP6_all_vs_data.csv")
na_ol_morning = pd.read_csv(filepath_or_buffer="/shared/suryansh.tiwari/output_csv/Redo_HI_FA_EP6_all_vs_data.csv")
```

```python
na_ol_night_ti = set(na_ol_night.Text_id)
na_ol_morning_ti = set(na_ol_morning.Text_id)
intersec = na_ol_night_ti.intersection(na_ol_night_ti)
diff = na_ol_night_ti.difference(na_ol_morning_ti)
```

```python
night = na_ol_night.loc[na_ol_night['Text_id'].isin(intersec)]
morning = na_ol_morning.loc[na_ol_morning['Text_id'].isin(intersec)]
```

```python
z_intersection = pd.merge(night, morning, on='Text_id', how='inner')
```

```python
z_intersection.head()
```

```python
z_intersection['match_text'] = z_intersection['Machine_Transcribed_text_x'] == z_intersection['Machine_Transcribed_text_y']
```

```python
z_mismatch = z_intersection.loc[z_intersection['match_text'] == False]
```
