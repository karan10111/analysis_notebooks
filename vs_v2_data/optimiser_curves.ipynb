```python
import numpy as np 
import seaborn as sns
import math
```

```python
global_steps = [x+1 for x in range(5000)]
init_lr = 2e-4
beta_1 = 0.9
beta_2 = 0.999
lr_ts = []
lr_t = init_lr
for step in global_steps:
    beta_1_power = math.pow(beta_1, step)
    beta_2_power = math.pow(beta_2, step)
    lr_t = init_lr * ((math.sqrt(1 - beta_2_power) / (1 - beta_1_power)))
    lr_ts.append(lr_t)
lr_t_fixed = np.array(lr_ts)
sns.lineplot(x=global_steps, y=lr_t_fixed)
```

```python
epochs= [x for x in range(10)]
init_lr = 2e-4
flag_decay = 1.5 # If relative improvement in loss after i-1 epoch is less that 5%, then flag_decay += 0.5
lr_ts = [init_lr/(((1+index)/2)**0.5) for index in epochs]
lr_t_actual = np.array([lr_t/(2 ** flag_decay) for lr_t in lr_ts])
```

```python
sns.lineplot(x=epochs, y=lr_t_actual)
```

```python
lr_t_actual
```

```python
total_epochs = 1000
epochs= [x for x in range(total_epochs)]
init_lr = 2e-4
flag_decay = 1.5
lr_ts = []
for index in epochs:
    lr_t = init_lr/(((1+index)/2)**0.5)
    lr_t_actual = lr_t/(2 ** flag_decay)
    lr_ts.append(lr_t_actual)
    flag_decay += 0.5
sns.lineplot(x=epochs, y=lr_ts)
```
