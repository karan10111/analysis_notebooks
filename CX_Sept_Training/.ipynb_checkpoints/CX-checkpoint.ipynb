# Computing Hours of Different Audio Files

## Tag 277

```python
# 0- Clean 1- Primary Und 2- Sec Non Prom 3- Sec Prom
fx_cx_data_277 = pd.read_sql("select * from data_transcribe where tag=277 and verified in (1,8,11,7,17,18);", db_connection)
fx_cx_data_277['corrected_utf_text_decoded'] = fx_cx_data_277['corrected_utf_text'].apply(base64.urlsafe_b64decode).str.decode('utf8')
fx_cx_data_277['type_of_text'] = fx_cx_data_277['corrected_utf_text_decoded'].apply(get_type_of_text)
fx_cx_data_277_counts = fx_cx_data_277['type_of_text'].value_counts()
ind_counts = (fx_cx_data_277.groupby(by="type_of_text")["length"].agg('sum'))/(1000*60*60)
ind_counts
```
